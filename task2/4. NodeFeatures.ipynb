## 3. Approach 2 – MLP (Node Features Only)

The MLP sees each node in isolation.  It can learn that "a paper about

### Architecture

$$
\hat{y} = \text{softmax}\!\left( \text{ReLU}(XW_1 + b_1) \cdot W_2 + b_2 \right)
$$

where $X \in \mathbb{R}^{N \times 1433}$ is the feature matrix and
$W_1, W_2$ are learned weight matrices.

## 4. Approach 3 – Vanilla GNN (Graph + Features)

A **Vanilla GNN** combines both views.  At each layer, every node
*aggregates* the feature vectors of its neighbours:

$$
H^{(k+1)} = \sigma\!\left( \hat{A} \, H^{(k)} \, W^{(k)} \right)
$$

where

* $\hat{A} = A + I$ – the adjacency matrix with **self-loops** added
  (so each node also aggregates its own representation).
* $H^{(k)}$ – node representations at layer $k$; $H^{(0)} = X$ (raw features).
* $W^{(k)}$ – a learned weight matrix (same for all nodes at layer $k$).
* $\sigma$ – a non-linearity (ReLU between layers, log-softmax at output).

After $K$ layers, each node's embedding encodes information from its
$K$-hop neighbourhood.

> **Compare with DeepWalk:** both exploit the graph.  But DeepWalk discards
> features; GNN *starts* from them and refines via neighbourhood aggregation.

> **Compare with MLP:** both use the same $W$ matrix applied to features.
> But MLP applies it once in isolation; GNN first aggregates neighbours, so
> the same features are shared and mixed across the graph.

### 5.3 Discussion

**Questions to think about:**

1. DeepWalk embeddings capture community structure well, but nodes with
   similar roles in different communities may be confused.  Which classes
   does it struggle with most in the t-SNE plot?

DeepWalk struggles most with classes that are highly dispersed across the graph rather than forming singular, isolated communities. In the t-SNE plot, this is particularly evident in the large red cluster, which is heavily "polluted" with nodes from other classes (like purple and brown). Because DeepWalk relies solely on local neighborhood structure (random walks), it maps structurally close nodes together even if their actual features/classes differ, and fails to group nodes of the same class if they reside in separate communities.

2. The MLP sees each node in isolation.  For Cora this works reasonably well
   because the bag-of-words features are already highly discriminative.  Can
   you think of a graph dataset where features would be *uninformative* and
   the MLP would fail completely?

A perfect example is a Molecular Graph dataset (e.g., predicting molecular properties or toxicity). In these graphs, nodes are atoms and node features are simply atom types (Carbon, Oxygen, Hydrogen, etc.). An MLP evaluates each node in isolation, meaning it only sees a "bag of atoms" without knowing how they are connected. Consequently, an MLP would fail completely when dealing with isomers—molecules that have the exact same count and type of atoms (identical features) but entirely different structural bonds and chemical properties. Without the graph topology, the MLP would output the exact same prediction for two functionally different molecules.

3. After two GNN layers, each node's embedding aggregates information from
   its **2-hop neighbourhood**.  What is the risk of adding more layers?
   (Hint: think about what $\hat{A}^K$ converges to as $K \to \infty$.)

The primary risk of adding too many GNN layers is a phenomenon known as over-smoothing. As $K \to \infty$, the aggregated matrix $\hat{A}^K$ converges to a stationary distribution (similar to the limit of a random walk). Mathematically, this means that the feature representations of all nodes will converge to the exact same global average vector. As a result, nodes lose their distinct local identities and become indistinguishable from one another, making it impossible for the model to perform node classification effectively.

4. Modify `build_adjacency` to also **normalise** each row by
   the node degree:
   $$\tilde{A} = D^{-1} \hat{A}$$
   where $D_{ii} = \sum_j \hat{A}_{ij}$.


   Does this improve accuracy?  Why or why not?

   Yes, it almost always improves accuracy and significantly stabilizes training.
   
   In a Vanilla GNN without normalization ($H^{(k+1)} = \hat{A} H^{(k)} W^{(k)}$), the message passing acts as a Sum Aggregation.If Node A has 1,000 neighbors and Node B has 2 neighbors, multiplying by $\hat{A}$ will result in Node A's feature vector containing the sum of 1,000 vectors, while Node B's vector is the sum of only 2.This causes the feature values of high-degree nodes to "explode" (numerical instability) and gradients to become massive. The network becomes biased towards popular nodes simply because their numbers are larger, not because their features are more meaningful.
   
   By applying $D^{-1} \hat{A}$, we are dividing each row by the number of connections. This changes the operation from a Sum Aggregation to a Mean Aggregation (averaging). Now, every node, regardless of whether it has 2 neighbors or 1,000, receives an aggregated feature vector that is on the exact same numerical scale.
